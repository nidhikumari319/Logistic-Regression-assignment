#12.How does class imbalance affect Logistic Regression.
ans:

    1.Biased Predictions → The model favors the majority
    class and ignores the minority class.
    2.Poor Recall for Minority Class → It may predict all
    cases as the majority class, leading to high False Negatives (FN).
    3.Misleading Accuracy → High accuracy but poor performance
    on the minority class (e.g., 95% accuracy when minority
    class is only 5%).
    4.Poor Decision Boundary → The model may not learn
    patterns for the minority class properly.

#15.How is Logistic Regression extended for multiclass classification.
ans:
   
    Logistic Regression is naturally used for binary
    classification (0 or 1), but it can be extended to handle
    multiple classes  using the following methods:

    1.One-vs-Rest (OvR) [Default in Scikit-Learn]:
    It is a method used to extend Logistic Regression to
    multiclass problems (3 or more classes).
    Instead of handling all classes at once, it breaks
    the problem into multiple binary classification problems.

    2.Multinomial Logistic Regression:
    It is an extension of Logistic Regression used for
    multiclass classification (3 or more classes).
    Instead of doing multiple binary classifications (like OvR),
    it predicts all classes at once using the Softmax function.

#20.How do we interpret coefficients in Logistic Regression?
ans:
   
    Each coefficient (β) shows how a feature affects the odds of the outcome:

    Positive (+β) → Increases the probability of class 1.
    Negative (−β) → Decreases the probability of class 1.
    Odds Ratio → e^β(how much the odds change for a one-unit increase in the feature).
    Intercept (β0) → Baseline probability when all features are zero.

###9.Write a Python program to apply RandomizedSearchCV for tuning hyperparameters (C, penalty, solver) in Logistic Regression. Print the best parameters and accuracy?

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load dataset from CSV file
df = pd.read_csv('sample_data.csv')  # Replace 'data.csv' with the actual filename
```

```python
# Assuming the last column is the target variable
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define Logistic Regression model
model = LogisticRegression()
```

```python
# Define parameter grid for RandomizedSearchCV
param_dist = {
    'C': np.logspace(-4, 4, 20),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}
```

```python
# Get the best parameters and accuracy
best_params = random_search.best_params_
print("Best Parameters:", best_params)
```

```python
# Evaluate the best model
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
# Find the optimal C
optimal_C = C_values[np.argmax(mean_accuracies)]
optimal_C
```
